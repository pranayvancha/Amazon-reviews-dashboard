```python
import nltk
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk import word_tokenize

def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('RB'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
def clean_text(text):
# pos tag text
    
    clean_t = []
    try:
        word_text = text.split()
    except:
        return text
    pos_tags = pos_tag(word_text)
    for text,pos in pos_tags:
        pos_tags = get_wordnet_pos(pos)
        if pos_tags == wordnet.ADJ or pos_tags == wordnet.ADV or pos_tags==wordnet.VERB :
            clean_t.append(text)
    result = ""
    for i in clean_t:
        result+= i+ " "
    return result
```

```python
data["clean_text_wordcloud"] = data["summary_clean"].apply(lambda x: clean_text(x))
```
